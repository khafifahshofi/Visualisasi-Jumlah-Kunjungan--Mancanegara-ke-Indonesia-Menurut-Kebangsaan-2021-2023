```python
from IPython import get_ipython
from IPython.display import display
# %%
from google.colab import files

uploaded = files.upload()  # Ini akan meminta Anda untuk mengunggah file dari komputer
# Get the filename of the uploaded file
file_path = list(uploaded.keys())[0]  # Assuming you uploaded only one file


# %%
import pandas as pd
import matplotlib.pyplot as plt

# %%
# file_path is now the correct filename
# Membaca sheet pertama dan membersihkan header
df = pd.read_excel(file_path, sheet_name=0, skiprows=2, names=["Kebangsaan", "2021", "2022", "2023"])

# Menghapus baris kosong
df.dropna(subset=["Kebangsaan"], inplace=True)

# Melihat 5 data teratas
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Menghapus kategori agregat seperti 'Total', 'ASEAN lainnya'
exclude_keywords = ["Total", "ASEAN lainnya"]
df_filtered = df[~df["Kebangsaan"].str.contains("|".join(exclude_keywords), case=False, na=False)]

# Memilih 10 negara dengan jumlah kunjungan tertinggi di 2023
top_countries = df_filtered.sort_values(by="2023", ascending=False).head(10)

# Plot garis untuk masing-masing negara
for _, row in top_countries.iterrows():
    plt.plot(["2021", "2022", "2023"], row[1:], marker="o", label=row["Kebangsaan"])

plt.title(" Tren Kunjungan Wisatawan Mancanegara ke Indonesia (2021-2023)", fontsize=14)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Jumlah Kunjungan", fontsize=12)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pastikan df sudah dibaca sebelumnya

# Menghapus kategori agregat seperti 'Total', 'ASEAN lainnya', dan variasi lainnya
exclude_keywords = ["Total", "ASEAN lainnya"]
df_filtered = df[~df["Kebangsaan"].str.contains("|".join(exclude_keywords), case=False, na=False)]

# Mengurutkan berdasarkan jumlah kunjungan di tahun 2023 dan mengambil 10 negara teratas
df_sorted = df_filtered.sort_values(by="2023", ascending=False).head(10)

# Membuat bar chart
plt.figure(figsize=(12,6))
sns.barplot(y=df_sorted["Kebangsaan"], x=df_sorted["2023"], hue=df_sorted["Kebangsaan"], palette="Blues_r", legend=False)

plt.xlabel("Jumlah Kunjungan", fontsize=12)
plt.ylabel("Kebangsaan", fontsize=12)
plt.title("10 Negara dengan Kunjungan Terbanyak ke Indonesia (2023)", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```
